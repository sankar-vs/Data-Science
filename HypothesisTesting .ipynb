- H0(Null Hypothesis) : Marriage age of IT professional Male in Bangalore 30
- Ha(Alternate Hypothesis) : Marriage age of IT professional Male in Bangalore is NOT 30
- Hypothotical Mean is 30

```python
ban_mar_age = [30,28,25,33,30,32,38,29,35,36,27]
```

# One Sample T-Test

```python
%matplotlib inline
import seaborn as sb
```

```python
from scipy import stats
```

```python
# ensure data is normal, T-Test are Parametric
print(stats.skew(ban_mar_age),stats.kurtosis(ban_mar_age))
sb.distplot(ban_mar_age)
stats.ttest_1samp(ban_mar_age,30)
```

- Prob Value, Pvalue = 35.7%, 
- Standard Rule: P < 5% Reject Null Hypothesis
- We failed to reject H0, so we accept H0
- We conclude that Marriage age in Bangalore of IT professional male is 30.

# Two Sample Independant T-Test

```python
ban_mar_age = [30,28,25,33,30,32,38,29,35,36,27]
hyd_mar_age = [28,27,28,29,31,26,27,33,29,27,26,24,29]
```

```python
print(stats.skew(ban_mar_age),stats.kurtosis(ban_mar_age))
sb.distplot(ban_mar_age)
stats.ttest_1samp(ban_mar_age,30)
```

- H0: Bangalore and Hyderabad marriage age of IT prof male are same
- Ha: Bangalore and Hyderabad marriage age of IT prof male are NOT same

```python
print(stats.skew(hyd_mar_age),stats.kurtosis(hyd_mar_age))
sb.distplot(hyd_mar_age)
stats.ttest_1samp(hyd_mar_age,30)
```

```python
stats.ttest_ind(ban_mar_age,hyd_mar_age)
```

- P = 2.3%, which is <5%, So we REJECT H0
- We conclude that Marriage age in Bangalore and Hyderabad of IT professional are NOT same.

# Relational T-Test
- Redbull Example

```python
before_rb = [45,56,39,52,64] # response time in centi-seconds 100centi-seconds = 1 seconds
after_rb = [25,36,29,32,24]
print(stats.skew(before_rb),stats.kurtosis(before_rb))
print(stats.skew(after_rb),stats.kurtosis(after_rb))
```

- H0 : Response time before and after Redbull are SAME
- Ha : Response time before and after Redbull are NOT SAME

```python
stats.ttest_rel(before_rb,after_rb)
```

## ANOVA (Analysis of Variance)

```python
lemon = [36,12,46,10,15,20]
coffee = [33,10,38,12,13,19]
rb = [31,9,39,13,10,15]
```

```python
print(stats.skew(lemon),stats.kurtosis(lemon))
print(stats.skew(coffee),stats.kurtosis(coffee))
print(stats.skew(rb),stats.kurtosis(rb))
```

- H0: Response times of all drinks are SAME
- Ha: Response times of all drinks are NOT SAME

```python
stats.f_oneway(lemon,coffee,rb)
```

- P = 88%, not <5%, we fail to reject H0, We accept H0
- Response times of all drinks are SAME
- Which means drinks have NO EFFECTS
